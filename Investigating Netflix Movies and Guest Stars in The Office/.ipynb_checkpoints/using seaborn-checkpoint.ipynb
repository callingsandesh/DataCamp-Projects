```python
# Import pandas and matplotlib.pyplot
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Read in the csv as a DataFrame
office_df = pd.read_csv('datasets/office_episodes.csv', parse_dates=['release_date'])
```

```python
office_df.head()
```

```python
# Initiatlize two empty lists
cols = []

# Iterate through the DataFrame, and assign colors based on the rating
for ind, row in office_df['scaled_ratings'].items():
    if row < 0.25:
        cols.append(1)
    elif row < 0.50:
        cols.append(3)
    elif row< 0.75:
        cols.append(5)
    else:
        cols.append(7)
```

```python
office_df['ratings_new']=cols
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.scatterplot(x='episode_number',y='viewership_mil',hue='ratings_new',size='ratings_new',palette="deep",sizes=(1, 500),style='has_guests',data=office_df)
ax.set_xlabel("episode_number",fontsize=24)
ax.set_ylabel("viewership in millions",fontsize=24)
# Create a title
plt.title("Popularity, Quality, and Guest Appearances on the Office", fontsize=28)
```
